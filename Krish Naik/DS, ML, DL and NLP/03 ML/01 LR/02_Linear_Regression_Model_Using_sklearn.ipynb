## Loading Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("./Data/data_for_lr.csv")
```

```python
data = data.dropna()
print(f"Shape of the data: {data.shape}")
```

```python
# Training input and output
# storing first 500 values from the column x and reshaping to maintain structure
train_input = np.array(data.x[0:500]).reshape(500, 1) 
train_output = np.array(data.y[0:500]).reshape(500, 1)

# Validation data input and output
test_input = np.array(data.x[500: 700]).reshape(199, 1) 
test_output = np.array(data.y[500: 699]).reshape(199, 1) # y coulmn is having a null value hence we have used 699
```

```python
print(f"training data input shape: {train_input.shape}")
print(f"training data output shape: {train_output.shape}")
print(f"test data input shape: {test_input.shape}")
print(f"test data output shape: {test_output.shape}")
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(train_input, train_output)
```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

## Predictions

```python
lr_predictions = lr_model.predict(test_input)
```

```python
plt.plot(test_input, test_output, "+", color = "red")
plt.plot(test_input, lr_predictions, "*", color = "green")
plt.title("Performace Testing")
plt.xlabel("input")
plt.ylabel("Output/Predictions")
plt.show()
```

```python
cost = mean_squared_error(test_output, lr_predictions)
cost
```
